```python
import numpy as np
import pandas as pd
```

```python
Xception = pd.read_csv("Xception.csv")
VGG = pd.read_csv("VGG.csv")
ResNet50 = pd.read_csv("ResNet50.csv")
DenseNet201 = pd.read_csv("DenseNet201.csv")
InceptionResNetV2 = pd.read_csv("InceptionResNetV2.csv")
```

```python
Majority_Vote = Xception
```

```python
Majority_Vote['VGG'] = VGG['class']
```

```python
Majority_Vote['ResNet50'] = ResNet50['class']
```

```python
Majority_Vote['DenseNet201'] = DenseNet201['class']
```

```python
Majority_Vote['InceptionResNetV2'] = InceptionResNetV2['class']
```

```python
Majority_Vote['Xception'] = Xception['class']
```

```python
Majority_Vote = Majority_Vote.drop(columns=["class"]).copy()
```

```python
Majority_Vote
```

```python
Majority_Vote.to_csv('Majority_Vote.csv', index = False)
```

```python
all_Majority_Vote = []
from scipy import stats
for i in range(len(Majority_Vote)) :
    bb = stats.mode(Majority_Vote.iloc[i,[1,2,3,4,5]])
    bb = bb[0][0]
    all_Majority_Vote.append(bb)
```

```python
all_Majority_Vote = pd.DataFrame(all_Majority_Vote , columns = ['class'])
Majority_Vote = pd.concat([Majority_Vote , all_Majority_Vote] , axis=1)
```

```python
Majority_Vote.to_csv('Majority_Vote.csv', index = False)
Majority_Vote
```

```python
Majority_Vote = Majority_Vote.drop(columns=["VGG"]).copy()
Majority_Vote = Majority_Vote.drop(columns=["ResNet50"]).copy()
Majority_Vote = Majority_Vote.drop(columns=["DenseNet201"]).copy()
Majority_Vote = Majority_Vote.drop(columns=["InceptionResNetV2"]).copy()
Majority_Vote = Majority_Vote.drop(columns=["Xception"]).copy()
```
